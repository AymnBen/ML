```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier

# Création données 
data = pd.DataFrame({
    'Age': [22, 25, 47, 52],
    'Revenu' :[2500, 3000, 9000, 11000],
    'Premium': [0,0,1,1]
})

# Séparation variables explicatives et variable cible
X = data[['Age', 'Revenu']]
y = data['Premium']

# Création et entraînement du modèle KNN
model = KNeighborsClassifier(n_neighbors=1)
model.fit(X, y)

# Nouveau client (Age: 30, Revenu: 4000)
nouveau_client = pd.DataFrame([[30, 4000]], columns=['Age', 'Revenu'])

# Prédiction 
prediction = model.predict(nouveau_client)


print('Prédection pour (30,4000) : ',prediction[0])
```

```python
# On importe la classe DecisionTreeClassifier de sklearn.tree 
from sklearn.tree import DecisionTreeClassifier

# On importe pandas pour la manipulation des données sous forme de tableau (DataFrame)
import pandas as pd

data = pd.DataFrame({
    'Heure': [1, 2, 6, 7],
    'Results': ['Echec', 'Echec', 'Reussite', 'Reussite']
})

# Affichage du dataset (optionnel pour vérifier que les données sont correctes)
print("Dataset : ")
print(data)

X = data[['Heure']]

# y = variable cible (ce que le modèle doit prédire)
# Ici le resultat (Echec ou Reussite)
y = data['Results']
# ------------------------------------
model = DecisionTreeClassifier(max_depth=1)

model.fit(X, y)


nouvelle_donnee = pd.DataFrame({'Heure': [5]})

prediction = model.predict(nouvelle_donnee)
# Affichage de la prédiction pour 5 heures
print('Prediction pour 5 heures : ', prediction[0])
```

```python
# Importation de l'outil qui transforme le texte en nombres (vectorisation)
from sklearn.feature_extraction.text import CountVectorizer

# Importation du modèle de classification Naive Bayes
from sklearn.naive_bayes import MultinomialNB


emails = ['offre gratuite', 'reunion projet']

# Labels associés à chaque email
# spam = email indésirable
# non_spam = email normal
labels = ['spam', 'non_spam']



vectorizer = CountVectorizer()
X = vectorizer.fit_transform(emails)


model = MultinomialNB()
model.fit(X, labels)



test_email = vectorizer.transform(['offre speciale'])

# Prédiction
prediction = model.predict(test_email)

# -----------------------------
# Affichage du résultat
# -----------------------------

print("Prédiction :", prediction[0])
```
