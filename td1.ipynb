```python
import pandas as pd 
import numpy as np 
data = pd.DataFrame({
    'Age': [25, 45, np.nan, 50, 23, 40],
    'Revenu' :[3000, 8000, 5000, np.nan, 2500, 7000],
    'Freaquence_achat': [1,5,3,6,1,4],
    'Ville': ['Rabat', 'Casablanca', 'Rabat', 'Fes', 'Fes', 'Casablanca'],
    'Client premium': [0,1,0,1,0,1]    
})
print(data)
```

```python
data["Age"] = data["Age"].fillna(data["Age"].mean())
data["Revenu"] = data["Revenu"].fillna(data["Revenu"].mean())
print(data)
```

```python
# data_encode = pd.get_dummies(data, columns=['Ville'])
data_encode = pd.get_dummies(data, columns=['Ville'], dtype=int)
print(data_encode)
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Séparation des variables
X = data_encode.drop(columns=['Client premium'])
y = data_encode['Client premium']

# Colonnes numériques à normaliser
num_cols = ['Age', 'Revenu', 'Freaquence_achat']

# Initialisation du scaler
scaler = StandardScaler()

# Application du scaler uniquement sur les colonnes numériques
X[num_cols] = scaler.fit_transform(X[num_cols])

# Affichage du résultat final
print(X)
```
